```python
import numpy as np
import matplotlib.pyplot as plt

n = 20
func = lambda x: x
noise = np.random.normal(0, n/10, size=n)

x = np.linspace(-10, 10, n)
x = np.round(x, 2)
y = func(x)
y_noisy = func(x) + noise
y_noisy = np.round(y_noisy, 2)

print(x)
print(y)

coeff = np.polyfit(x, y_noisy, 2)
fit_func = np.poly1d(coeff)
print(fit_func)

plt.scatter(x, y_noisy)
plt.plot(x, fit_func(x))
plt.plot(x, y)
```
