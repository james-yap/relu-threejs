```python
import numpy as np
import matplotlib.pyplot as plt
import pyperclip

np.set_printoptions(precision=2, floatmode='fixed')

n = 10
func = lambda x: x
noise = np.random.normal(0, 0.3, size=n)

x = np.linspace(0, 3, n)
# x = np.round(x, 2)
y = func(x)
y_noisy = func(x) + noise
# y_noisy = np.round(y_noisy, 2)

res = f"""const x = {x.tolist()};
const y = {y_noisy.tolist()};
"""

pyperclip.copy(res)
print(res)

coeff = np.polyfit(x, y_noisy, 1)
fit_func = np.poly1d(coeff)
print(coeff)

plt.scatter(x, y_noisy)
plt.plot(x, fit_func(x))
plt.plot(x, y)
```

```python
n = 10
def func(x):
    return (x - 1.5) ** 2
noise = np.random.normal(0, 0.1, size=n)

x = np.linspace(0, 3, n)
y = func(x)
y_noisy = func(x) + noise

res = f"""const x = {x.tolist()};
const y = {y_noisy.tolist()};
"""

pyperclip.copy(res)
print(res)

coeff = np.polyfit(x, y_noisy, 2)
fit_func = np.poly1d(coeff)
print(coeff)

plt.scatter(x, y_noisy)
plt.plot(x, fit_func(x))
plt.plot(x, y)
```
